```python
#read a lemma dictionary
#lemma  wordform
def readDictionary(dictionary) :
    Dic = {}
    with open(dictionary,"r", encoding="utf8") as file:
        for entry in file:
#            print(entry, flush=True, end = '')
            lem, tok = re.findall("^(.*?)[\s]+(.*?)$", entry)[0]
            Dic[tok.lower()] = lem.lower()
    return (Dic)


#read a lemma dictionary
#lemma -> <POS1>word1,<POS2>word2
def readDictionaryPos(dictionary, verbose = 0) :
    Dic = {}
    with open(dictionary,"r", encoding="utf8") as file:
        for entry in file:
            entry = entry.replace('\n', '') 
            if verbose : print("entry", entry, flush=True)
            l = re.search(f"^([\w]+)", entry) #lemma
            if not l: continue
                
            l1 = str(l.group()).lower()
            E = re.findall(r"<([^>]+)>([^,]+)", entry)
            for p, w in list(E):
                w1 = str(w).lower()
                if verbose: print("\tparse", l1, p, w1, flush=True)
                Dic.setdefault(w1,{})
                Dic[w1][p] = l1
                #something efficient 'yes'
                
                # 'VVG': 'ignore', 'VBG': 'ignore'}
                if (p == "VVG") : Dic[w1]["VBG"] = l1
                if (p == "VBG") : Dic[w1]["VVG"] = l1
                    
                # VBD': 'ignore', VVD': 'ignore', 'VVN': 'ignore', 'VBN': 'ignore'
                if (p == "VVD") : Dic[w1]["VVN"] = l1
                if (p == "VVD") : Dic[w1]["VBN"] = l1
                if (p == "VVD") : Dic[w1]["VBD"] = l1
                    
                if (p == "VVN") : Dic[w1]["VVD"] = l1
                if (p == "VVN") : Dic[w1]["VBN"] = l1
                if (p == "VVN") : Dic[w1]["VBB"] = l1
                    
                if (p == "VBN") : Dic[w1]["VVD"] = l1
                if (p == "VBN") : Dic[w1]["VVN"] = l1
                if (p == "VBN") : Dic[w1]["VBD"] = l1
                    
                if (p == "VBD") : Dic[w1]["VVD"] = l1
                if (p == "VBD") : Dic[w1]["VVN"] = l1
                if (p == "VBD") : Dic[w1]["VBN"] = l1
                    
                #VBZ 'VVZ': 'want'}
                if (p == "VBZ") : Dic[w1]["VVZ"] = l1
                if (p == "VVZ") : Dic[w1]["VBZ"] = l1
                
    return (Dic)
```

```python
# Stemmer with Pos
def myStemmer2(wrd, pos) :  
        
    stem = wrd.lower()
                                 
    if(pos == "VBD"):
        stem = re.sub(r'(ed)$', '', stem)

    elif(pos == "VBN"):
        stem = re.sub(r'(ed)$', '', stem)
        
    elif(pos == "VBG"):
        stem = re.sub(r'(ing)$', '', stem)             
        
    elif(pos == "VBZ"):
        stem = re.sub(r'(s)$', '', stem)
     
    elif(pos == "NN"):
        stem = re.sub(r'(ment|ence|ation)$', '', stem)

    elif(pos == "NNS"):
        stem = re.sub(r'(s|ies)$', '', stem)
        
    elif(pos == "NNP"):
        stem = re.sub(r'(ed)$', '', stem)
        
    elif(pos =="JJ"):
        if(re.search(r'[sl]ing', stem)) : 
            stem = re.sub(r'[sl]ing$', '', stem)
        else : 
            stem = re.sub(r'(est|ing|ly)$', '', stem)

    elif(pos == "JJR"):
        stem = re.sub(r'(er)$', '', stem)               
        
    elif(pos == "JJS"):
        stem = re.sub(r'(est)$', '', stem)
        
    elif(pos == "RBR"):
        stem = re.sub(r'(er)$', '', stem)               
        
    elif(pos == "RBS"):
        stem = re.sub(r'(est)$', '', stem)                
        
    elif(pos == "RB"):
        stem = re.sub(r'(ly|ed)$', '', stem)    
        
    elif(pos == "FW"):
        stem = re.sub(r'(o)$', '', stem)                
    
    return(stem)
```

```python
#task:
# amend myStemmer2 to beat porter stemmer

# Stemmer with Pos
def myStemmer2(wrd, pos) :  
        
    stem = wrd.lower()
                                 
    if(pos == "VBD"):
        if(re.search(r'(used)$', stem)) : 
            stem = re.sub(r'(used)$', 'use', stem)
        elif(re.search(r'(llied)$', stem)) : 
            stem = re.sub(r'(llied)$', 'lly', stem)
        else : 
            stem = re.sub(r'(ed)$', '', stem)
            
    elif(pos == "VBN"):
        if(re.search(r'(used)$', stem)) : 
            stem = re.sub(r'(used)$', 'use', stem)
        elif(re.search(r'(llied)$', stem)) : 
            stem = re.sub(r'(llied)$', 'lly', stem)
        else : 
            stem = re.sub(r'(ed)$', '', stem)
        
    elif(pos == "VBG"):
        stem = re.sub(r'(ing)$', '', stem)
        
    elif(pos == "VBZ"):
        if(re.search(r'(is)$', stem)) : 
            stem = re.sub(r'(is)$', 'be', stem)
        elif(re.search(r'(ies)$', stem)) : 
            stem = re.sub(r'(ies)$', 'y', stem)
        elif(re.search(r'(\s s)$', stem)) : 
            stem = re.sub(r'(s)$', 's', stem)
        else :
            stem = re.sub(r'(s)$', '', stem)
     
    elif(pos == "NN"):
        stem = re.sub(r'(ment|ence|ation)$', '', stem)
    
    elif(pos == "NN$"):
        stem = re.sub(r'.*\'s$', '', stem)

    elif(pos == "NNS"):
        stem = re.sub(r'(s|ies)$', '', stem)
        
    elif(pos == "NNP"):
        stem = re.sub(r'(ed)$', '', stem)
        
    elif(pos =="JJ"):
        if(re.search(r'[sl]ing', stem)) : 
            stem = re.sub(r'[sl]ing$', '', stem)
        else : 
            stem = re.sub(r'(est|ing|ly)$', '', stem)

    elif(pos == "JJR"):
        stem = re.sub(r'(er)$', '', stem)               
        
    elif(pos == "JJS"):
        stem = re.sub(r'(est)$', '', stem)
        
    elif(pos == "RBR"):
        stem = re.sub(r'(er)$', '', stem)               
        
    elif(pos == "RBS"):
        stem = re.sub(r'(est)$', '', stem)                
        
    elif(pos == "RB"):
        stem = re.sub(r'(ly|ed)$', '', stem)    
        
    elif(pos == "FW"):
        stem = re.sub(r'(o)$', '', stem) 
        
    elif(pos == "CD"):
        stem = re.sub(r'^-?[0-9]+(.[0-9]+)?$', '', stem)
    
    return(stem)
```
